## 実験結果


```python
%matplotlib inline
import os.path as osp
plt.style.use('ggplot')

# exp3 directory
exp_id = 'exp3'
exp_dir = osp.join('./records', exp_id)
```

### 結果1. 学習曲線

```python
fn_progress = './progress.csv'
df = pd.read_csv(osp.join(exp_dir, fn_progress))

plt.rcParams["font.size"] = 14
a0 = df.loc[df.AgentID==0, ['EpRewMean']].values
a1 = df.loc[df.AgentID==1, ['EpRewMean']].values
plt.plot(a0, label='Agent0', alpha=1, color='red')
plt.plot(a1, label='Agnet1', alpha=0.8, color='turquoise')
plt.legend(loc='best')
plt.ylim(-1500, 1100)
plt.axhline(1000,  ls='--', color='blue')
plt.axhline(-1000,  ls='--', color='blue')
plt.ylabel('reward')
plt.xlabel('iteration')
```

### 結果2. 実際の試合（アニメーション; 異なるiterの学習モデルでデモ）

```python
from IPython.display import HTML
def play_video(exp_id, iters, seed):
    par_list = iters + [seed]
    results_video = osp.join('./records', exp_id, 'models_%d_%d_seed%d.mp4' % tuple(par_list))
    return HTML("""
    <video width="500" height="300" controls>
      <source src="%s" type="video/mp4">
    </video>""" % results_video)
```

### Agent0 (iter=800) vs Agent1 (iter=800)


```python
play_video(exp_id, [800, 800], 10)
```

### Agent0 (iter=1920) vs Agent1 (iter=1920)
both are the latest models

```python
play_video(exp_id, [1920, 1920], 10)
```
