## 実験結果

```python
%matplotlib inline
import os.path as osp
plt.style.use('ggplot')

# exp2 directory
exp_id = 'exp_run-to-goal-ants-v0_2017-12-24_14-34-38'
exp_dir = osp.join('./records', exp_id)
```

### 結果1. 学習曲線

```python
fn_progress = './progress.csv'
df = pd.read_csv(osp.join(exp_dir, fn_progress))

plt.rcParams["font.size"] = 14
a0 = df.loc[df.AgentID==0, ['EpRewMean']].values
a1 = df.loc[df.AgentID==1, ['EpRewMean']].values
plt.plot(a0, label='Agent0', alpha=1, color='red')
plt.plot(a1, label='Agnet1', alpha=0.8, color='turquoise')
plt.legend(loc='best')
plt.ylabel('reward')
plt.xlabel('iteration')
```

### 結果2. 最新モデル(Agent0) vs 過去のモデル（Agent1)

```python
fn_count = osp.join('./records', exp_id, 'count_win_defeat.csv')
df = pd.read_csv(fn_count)

plt.rcParams["font.size"] = 14
iters = df['iter_agent1'].values
win_rates = ((df['win0'] + 0.5 * df['draw']) / df[['win0', 'win1', 'draw']].sum(axis=1)).values
errors = (np.sqrt(df['win0'] + 0.5 * df['draw']) / df[['win0', 'win1', 'draw']].sum(axis=1)).values
plt.errorbar(iters, 100*win_rates, yerr=100*errors, fmt='-o', capsize=0, label='Agent0: 2035 iters')
plt.axhline(50,  ls='--', color='blue')
plt.axhline(100,  ls='--', color='green')
plt.xlim([-10, 1000])
plt.ylabel('win rate (%)')
plt.xlabel("Agent1-model iteration")
plt.legend()
```

### 結果3. 実際の試合（アニメーション; 異なるiterの学習モデルでデモ）

```python
from IPython.display import HTML
def play_video(exp_id, iters):
    par_list = iters
    results_video = osp.join('./records', exp_id, 'models_%d_%d.mp4' % tuple(par_list))
    return HTML("""
    <video width="500" height="300" controls>
      <source src="%s" type="video/mp4">
    </video>""" % results_video)
```
